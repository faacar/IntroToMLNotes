```python
import pandas as pd
import numpy as np

data = pd.read_csv('/content/kaggle_data/pokemon.csv')
data.head()
```

```python
series = data['Defense']
print(type(series))

data_frame = data[['Defense']]
print(type(data_frame))
```

```python
defense_filter = data['Defense'] > 200

data[defense_filter]
```

```python
#AND logical operator
data[np.logical_and(data['Defense'] > 200, data['Attack'] > 100)]
#takes two parameters and return array
```

```python
#same result with the previous one 
data[(data['Defense'] > 200) & (data['Attack'] > 100)]
```

```python
#NumPy Quick Tutorial
#create array
one_dim_arr = np.array([1.2,2.2,3.8])
print(one_dim_arr)
print('-------')
two_dim_arr = np.array([[6,5],[8,7],[11,9]])
print(two_dim_arr)
print('-------')
# generate sequence between two parameters, first parameter included last one not
sequence_of_integers = np.arange(5,12)
print(sequence_of_integers)
print('-------')
#generate random number between two paramters. Also want size parameters
rand_int_bt50_100 = np.random.randint(low=50, high=101, size = 6)
print(rand_int_bt50_100)
print('-------')

#create random float number
#only takes number of generating values
random_floats_num = np.random.random([6])
print(random_floats_num)
print('-------')

#math operations on NumPy
random_floats_between_2_and_3 = random_floats_num + 2.0
print(random_floats_between_2_and_3)
```
